```scala
// Mutable variables using `var`
var greeting: String = "Welcome"
println(greeting)
greeting = "Goodbye"
println(greeting)
```

```scala
// Immutable variables using `val`
val number: Int = 10
println(number)
number = 20 // This would cause a compilation error
```

```scala
// Examples of integer literals
val decimal: Int = 42         // Decimal literal
val hex: Int = 0x2A           // Hexadecimal literal
val binary: Int = Integer.parseInt("101010", 2) // Binary literal (converted from binary string)
val longNumber: Long = 123456789L // Long literal
println(s"Decimal: $decimal, Hex: $hex, Binary: $binary, Long: $longNumber")
```

```scala
// Examples of floating-point literals
val pi: Float = 3.14f       // Float literal
val scientific: Double = 1.2e-5  // Scientific notation (Double)
val shortForm: Float = .5f  // Short form
val doubleVal: Double = 10.0
println(s"Pi: $pi, Scientific: $scientific, Short Form: $shortForm, Double: $doubleVal")
```

```scala
// Examples of Boolean literals
val isActive: Boolean = true
val isComplete: Boolean = false
println(s"Is Active: $isActive, Is Complete: $isComplete")
```

```scala
// Examples of Character literals
val letter: Char = 'X'
val newline: Char = '\n'
println(s"Letter: $letter, Newline: ${newline}Here")
```

```scala
// Examples of String literals
val simpleString: String = "Scala is great!"
val escapedString: String = "This is a \"quoted\" string."
val multilineString: String = """This string spans
multiple lines."""
println(simpleString)
println(escapedString)
println(multilineString)
```

```scala
// Null value examples
val nullableString: String = null
val nullableInt: Integer = null
println(s"Nullable String: $nullableString, Nullable Int: $nullableInt")
// Output:
// Nullable String: null, Nullable Int: null
```
