```scala
// If Statement
val temperature = 15

if (temperature >= 20) {
  println("It's a warm day.")
} else {
  println("It's a cold day.")
}
```

```scala
// If-Else Chained
val grade = 75

if (grade >= 90) {
  println("Excellent!")
} else if (grade >= 75) {
  println("Good Job!")
} else if (grade >= 50) {
  println("Needs Improvement.")
} else {
  println("Fail.")
}
```

```scala
val language = "Scala"

language match {
  case "Java" => println("The language is Java.")
  case "Scala" => println("The language is Scala.")
  case "Python" => println("The language is Python.")
  case _ => println("Unknown language.")
}
```

```scala
// Return Statement
def multiply(a: Int, b: Int): Int = {
  return a * b // Returning the product
}

val product = multiply(6, 7)
println(s"Product: $product")
```

```scala
import scala.util.control.Breaks._

breakable {
  for (i <- 1 to 10) {
    if (i == 7) break() // Exit the loop when i is 7
    println(s"Current number: $i")
  }
}
```

```scala
// Do-While Loop
var counter = 3
do {
  println(s"Counter value: $counter")
  counter -= 1
} while (counter > 0)
```
